# program on ATM

```python
import time # to wait for small time , like seconds
    
pin = int(input('enter your pincode: '))
if pin == 1234:
    print('welcome to the statebank')
    print('1.withdraw')
    print('2.transfer money')
    print('3.pin change')
    print('4.balance enquiry')
    
    choice = int(input('enter your choice here: '))
    
    if choice ==1:
        print('1.savings')
        
        per = int(input('enter your performence: '))
        if per == 1:
            cash = int(input('enter cash below: '))
            print('wait....')
            time.sleep(2)
            bal = 40000-cash
            print('balance left: ',bal)
            
    elif choice == 2:
        name = input('enter the name of reciever ')
        acc = int(input('enter account number: '))
        
        if acc == 123456789 and name == 'vallela':
            cas = int(input('enter the amount'))
            time.sleep(3)
            print('transaction completed')
            
        else:
            print('timeout')
            
    elif choice == 3:
        old = int(input('enter the old pass code: '))
        new = int(input('enter the new pass code: '))
        time.sleep(4)
        print('pin code changed')
        
    elif choice == 4:
        print('balance left in your account')
        time.sleep(5)
        print('40000/- only')
        print('thank you for your service')
```
